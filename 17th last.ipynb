```python
#k means clustering
#k means clustering is an unsupervised learning algorithm that will attempt to group similar clusters together from your data.

#it is mainly used in

#clustering similar documents
#clustering customers based on similar features
```

```python
from sklearn.datasets import make_blobs
#The make_blobs functions is used to generate synthetic datasets for clustering and clssification tasks
#This function will create clusters of data points with gaussian distribution.
```

```python
#creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
```

```python
#n_sample=Total number of points equally divided among clusters
#n_features=It indicated  the number of features (columns)
#centers = it determines no of clusters to be generated 
#cluster_std = it sets the standers deviation of the clusters ,High value makes the clusters to be spread out 
data
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:,0],x[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatters plot for k means")
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatters plot for k means")
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow',edgecolor='black',s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatters plot for k means")
```

```python
#to name the column
df_feat=pd.DataFrame(Scaled_features,columns=df.columns[:-1])
df_feat
```

```python
df_feat.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_feat
y=df['TARGET CLASS']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
KNN
```

```python
#to train model
KNN.fit(x_train,y_train)
```

```python
pred=KNN.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred,y_test)
acc
```

```python
#to find error rate
error_rate=[]
for val in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=val)
    knn.fit(x_train,y_train)
    pred_i=knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
error_rate
```

```python
plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate VS value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```
